Operaciones basicas

```python
G = nx.Graph()
G.add_edge(1,2)
G.add_edge(2,3, weight=0.9)
G.add_edge("A", "B")
G.add_edge("B", "B")
G.add_node("C")
G.add_node(print)

nx.draw_spring(G, with_labels=True)
plt.show()
```

Dibujos distintos

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G=nx.Graph()
G.add_edges_from(edge_list)
nx.draw_spring(G, with_labels=True)
plt.show()


nx.draw_circular(G, with_labels=True)
plt.show()


nx.draw_shell(G, with_labels=True)
plt.show()


nx.draw_spectral(G, with_labels=True)
plt.show()

nx.draw_random(G, with_labels=True)
plt.show()

nx.draw_planar(G, with_labels=True)
plt.show()
```

Grado

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G=nx.Graph()
G.add_edges_from(edge_list)
print(dict(G.degree)[5]) #Grado del vertice 3
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G=nx.DiGraph()
G.add_edges_from(edge_list)
print(dict(G.degree)[3]) #Grado del vertice
print(dict(G.in_degree)[3]) #Grado del vertice entrada incoming
print(dict(G.out_degree)[3]) #Grado del vertice salida outgoing
nx.draw_spring(G, with_labels=True)
plt.show()
```

Shortest path

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7), (2,8),(8,9),(9,4)]
G=nx.Graph()
G.add_edges_from(edge_list)
print(nx.shortest_path(G,2,4))
nx.draw_planar(G, with_labels = True)
```

Centralidad

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7), (2,8),(8,9),(9,4)]
G=nx.Graph()
G.add_edges_from(edge_list)

print(nx.closeness_centrality(G))
print("###")
print(nx.eigenvector_centrality(G))
print("###")
print(nx.degree_centrality(G))
print("###")

nx.draw_planar(G, with_labels = True)
plt.show()
```

Relabel

```python
G1 = nx.complete_graph(5)
G2 = nx.complete_graph(5)
G2 = nx.relabel_nodes(G2, {0:"A", 1:"B", 2:"C", 3:"D", 4:"E"})
G_connector = nx.from_edgelist([(4,"X"), ("X","A")])

G = nx.compose_all([G1,G2,G_connector])

print(nx.density(G))
print(nx.diameter(G))


nx.draw_spring(G, with_labels = True)
plt.show()
```

Eulerian path

```python
edge_list = [(1,2),(2,3),(3,4),(2,8),(8,9),(9,4)]
G=nx.Graph()
G.add_edges_from(edge_list)

print(list(nx.eulerian_path(G))) #all edges one time
print(list(nx.find_cliques(G))) #Subgraphs

nx.draw_planar(G, with_labels = True)
plt.show()
```

Connected component

```python
G1 = nx.complete_graph(5)
G2 = nx.complete_graph(5)
G2 = nx.relabel_nodes(G2, {0:"A", 1:"B", 2:"C", 3:"D", 4:"E"})


G = nx.compose_all([G1,G2])

print(list(nx.connected_components(G)))

nx.draw_spring(G, with_labels = True)
plt.show()
```
